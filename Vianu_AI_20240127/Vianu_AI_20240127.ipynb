**Linear Regression**

Regresia liniară este o metodă statistică care este utilizată pentru a modela relația dintre o variabilă dependentă (y) și o variabilă independentă (x). Modelul liniar presupune că există o relație liniară între cele două variabile, ceea ce înseamnă că putem reprezenta relația cu o ecuație a unei linii drepte.

De exemplu, dacă vrem să modelăm relația dintre venitul unei persoane (y) și numărul de ani de experiență (x), putem folosi ecuația:

`y = a * x + b`, unde:
* a este coeficientul de regresie al variabilei independente (x)
* b este termenul constant

Gradient descent este un algoritm de optimizare care este utilizat pentru a găsi valorile pentru coeficienții de regresie a modelului liniar. Gradient descent funcționează prin calcularea gradientului unei funcție de eroare, care este o măsură a cât de departe sunt predicțiile modelului de valorile reale. Apoi, gradientul este utilizat pentru a actualiza valorile coeficienților de regresie într-o direcție care reduce eroarea.

Eroarea este calculată ca suma pătratelor diferențelor dintre valorile predictate ale variabilei dependente (y) și valorile reale ale variabilei dependente. Cu cât diferențele sunt mai mici, cu atât eroarea este mai mică.

Gradient descent este un algoritm iterativ, ceea ce înseamnă că se repetă până când eroarea devine suficient de mică.

````python
import pandas as pd
from sklearn.linear_model import LinearRegression

# create a dataframe
df = pd.DataFrame({'x': [1, 2, 3, 4, 5], 'y': [2, 4, 5, 4, 5]})

# create a linear regression object
lr = LinearRegression()

# fit the model to the data
lr.fit(df[['x']], df['y'])

# make predictions
y_pred = lr.predict(df[['x']])

# print the predictions
print(y_pred)
````

````python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# create a dataframe
df = pd.DataFrame({'x': [1, 2, 3, 4, 5], 'y': [2, 4, 5, 4, 5]})

# create a linear regression object
lr = LinearRegression()

# fit the model to the data
lr.fit(df[['x']], df['y'])

# make predictions
y_pred = lr.predict(df[['x']])

# print the predictions
print(y_pred)

# plot the data
plt.scatter(df['x'], df['y'])

# plot the fitted line
plt.plot(df['x'], y_pred, color='red')

# label the axes
plt.xlabel('x')
plt.ylabel('y')

# show the plot
plt.show()
````

Regresia liniară este o metodă statistică utilizată pentru a modela relația dintre o variabilă dependentă (y) și o variabilă independentă (x).

Modelul de regresie liniară presupune că relația dintre cele două variabile este liniară, ceea ce înseamnă că putem reprezenta relația cu o ecuație a unei linii drepte. De exemplu, dacă vrem să modelăm relația dintre venitul unei persoane (y) și numărul de ani de experiență (x), putem folosi ecuația:

y = a * x + b
unde a este coeficientul de regresie al variabilei independente (x) și b este termenul constant.

Din punct de vedere matematic, regresia liniară este potrivită pentru a fi utilizată atunci când îndeplinește următoarele condiții:

Relația dintre variabile este liniară sau aproape liniară. Aceasta înseamnă că relația dintre variabile poate fi reprezentată în mod adecvat de o ecuație a unei linii drepte.

Variabila independentă (x) este măsurabilă cu precizie. Aceasta înseamnă că variabilele independente trebuie să fie cuantificabile și să poată fi măsurate cu o anumită precizie.

Variabila dependentă (y) este continuă. Aceasta înseamnă că variabila dependentă trebuie să fie un număr real și să poată lua orice valoare reală.

````python
import pandas as pd
import numpy as np

# create a dataframe
df = pd.DataFrame({'x': [1, 2, 3, 4, 5], 'y': [2, 4, 5, 4, 5]})

# fit the model to the data
def fit_linear_regression(X, y):
    n = len(X)
    theta_hat = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y))
    return theta_hat

theta_hat = fit_linear_regression(df[['x']], df['y'])

# make predictions
def predict_linear_regression(X, theta):
    y_hat = np.dot(X, theta)
    return y_hat

y_pred = predict_linear_regression(df[['x']], theta_hat)

# print the predictions
print(y_pred)
````

```X este matricea de date care conține variabilele independente.```

```y este vectorul de date care conține variabilele dependente.```

```n este numărul de observații în datele de antrenament.```

```theta_hat este vectorul de parametri ai modelului de regresie liniară.```

```np.linalg.inv(np.dot(X.T, X)) calculează inversa matricei X^TX.```

```np.dot(X.T, y) calculează produsul scalar X^Ty.```

```np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y)) calculează theta_hat.```

```y_hat este vectorul cu predicțiile pentru noile date.```

```np.dot(X, theta) calculează produsul scalar X * theta.```

Linear regression is a statistical approach to find and determine a relationship among an independent variable `x` and a dependent variable `y`. For us, our `x` is `Age` while our `y` is `Selling_Price`. In the below equation, linear regression helps us find the `m` and `b` that best relates our variables. In other words, we are trying to find a "line of best fit" for our variables `x` and `y`.

$y= mx + b$

[![Line of Best Fit](https://upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Linear_regression.svg/400px-Linear_regression.svg.png)](https://upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Linear_regression.svg/400px-Linear_regression.svg.png)

````python
import sklearn
from sklearn.linear_model import LinearRegression
import numpy as np

X = car_data[['Age']]
y = car_data[['Selling_Price']]
````

````python
print(X)
````

````python
print(y)
````

````python
# set up our model
linear = LinearRegression()

# train the model
linear.fit(X, y)
````

````python
import matplotlib.pyplot as plt

y_pred = linear.predict(X)
plt.plot(X, y_pred, color='red')

plt.scatter(X, y)
plt.xlabel('Age')
plt.ylabel('Selling_Price (lakhs)')
plt.show()
````

Mean squared error (MSE): This is the most common metric for evaluating the performance of linear regression models. It is calculated by averaging the squared differences between the predicted values and the actual values.

````python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_pred, y)
print("MSE:", mse)
````
