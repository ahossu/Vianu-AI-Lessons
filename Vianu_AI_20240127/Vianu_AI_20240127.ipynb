**Linear Regression**

Regresia liniar este o metod statistic care este utilizat pentru a modela relaia dintre o variabil dependent (y) i o variabil independent (x). Modelul liniar presupune c exist o relaie liniar ntre cele dou variabile, ceea ce nseamn c putem reprezenta relaia cu o ecuaie a unei linii drepte.

De exemplu, dac vrem s modelm relaia dintre venitul unei persoane (y) i numrul de ani de experien (x), putem folosi ecuaia:

`y = a * x + b`, unde:
* a este coeficientul de regresie al variabilei independente (x)
* b este termenul constant

Gradient descent este un algoritm de optimizare care este utilizat pentru a gsi valorile pentru coeficienii de regresie a modelului liniar. Gradient descent funcioneaz prin calcularea gradientului unei funcie de eroare, care este o msur a ct de departe sunt prediciile modelului de valorile reale. Apoi, gradientul este utilizat pentru a actualiza valorile coeficienilor de regresie ntr-o direcie care reduce eroarea.

Eroarea este calculat ca suma ptratelor diferenelor dintre valorile predictate ale variabilei dependente (y) i valorile reale ale variabilei dependente. Cu ct diferenele sunt mai mici, cu att eroarea este mai mic.

Gradient descent este un algoritm iterativ, ceea ce nseamn c se repet pn cnd eroarea devine suficient de mic.

````python
import pandas as pd
from sklearn.linear_model import LinearRegression

# create a dataframe
df = pd.DataFrame({'x': [1, 2, 3, 4, 5], 'y': [2, 4, 5, 4, 5]})

# create a linear regression object
lr = LinearRegression()

# fit the model to the data
lr.fit(df[['x']], df['y'])

# make predictions
y_pred = lr.predict(df[['x']])

# print the predictions
print(y_pred)
````

````python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# create a dataframe
df = pd.DataFrame({'x': [1, 2, 3, 4, 5], 'y': [2, 4, 5, 4, 5]})

# create a linear regression object
lr = LinearRegression()

# fit the model to the data
lr.fit(df[['x']], df['y'])

# make predictions
y_pred = lr.predict(df[['x']])

# print the predictions
print(y_pred)

# plot the data
plt.scatter(df['x'], df['y'])

# plot the fitted line
plt.plot(df['x'], y_pred, color='red')

# label the axes
plt.xlabel('x')
plt.ylabel('y')

# show the plot
plt.show()
````

Regresia liniar este o metod statistic utilizat pentru a modela relaia dintre o variabil dependent (y) i o variabil independent (x).

Modelul de regresie liniar presupune c relaia dintre cele dou variabile este liniar, ceea ce nseamn c putem reprezenta relaia cu o ecuaie a unei linii drepte. De exemplu, dac vrem s modelm relaia dintre venitul unei persoane (y) i numrul de ani de experien (x), putem folosi ecuaia:

y = a * x + b
unde a este coeficientul de regresie al variabilei independente (x) i b este termenul constant.

Din punct de vedere matematic, regresia liniar este potrivit pentru a fi utilizat atunci cnd ndeplinete urmtoarele condiii:

Relaia dintre variabile este liniar sau aproape liniar. Aceasta nseamn c relaia dintre variabile poate fi reprezentat n mod adecvat de o ecuaie a unei linii drepte.

Variabila independent (x) este msurabil cu precizie. Aceasta nseamn c variabilele independente trebuie s fie cuantificabile i s poat fi msurate cu o anumit precizie.

Variabila dependent (y) este continu. Aceasta nseamn c variabila dependent trebuie s fie un numr real i s poat lua orice valoare real.

````python
import pandas as pd
import numpy as np

# create a dataframe
df = pd.DataFrame({'x': [1, 2, 3, 4, 5], 'y': [2, 4, 5, 4, 5]})

# fit the model to the data
def fit_linear_regression(X, y):
    n = len(X)
    theta_hat = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y))
    return theta_hat

theta_hat = fit_linear_regression(df[['x']], df['y'])

# make predictions
def predict_linear_regression(X, theta):
    y_hat = np.dot(X, theta)
    return y_hat

y_pred = predict_linear_regression(df[['x']], theta_hat)

# print the predictions
print(y_pred)
````

```X este matricea de date care conine variabilele independente.```

```y este vectorul de date care conine variabilele dependente.```

```n este numrul de observaii n datele de antrenament.```

```theta_hat este vectorul de parametri ai modelului de regresie liniar.```

```np.linalg.inv(np.dot(X.T, X)) calculeaz inversa matricei X^TX.```

```np.dot(X.T, y) calculeaz produsul scalar X^Ty.```

```np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y)) calculeaz theta_hat.```

```y_hat este vectorul cu prediciile pentru noile date.```

```np.dot(X, theta) calculeaz produsul scalar X * theta.```

Linear regression is a statistical approach to find and determine a relationship among an independent variable `x` and a dependent variable `y`. For us, our `x` is `Age` while our `y` is `Selling_Price`. In the below equation, linear regression helps us find the `m` and `b` that best relates our variables. In other words, we are trying to find a "line of best fit" for our variables `x` and `y`.

$y= mx + b$

[![Line of Best Fit](https://upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Linear_regression.svg/400px-Linear_regression.svg.png)](https://upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Linear_regression.svg/400px-Linear_regression.svg.png)

````python
import sklearn
from sklearn.linear_model import LinearRegression
import numpy as np

X = car_data[['Age']]
y = car_data[['Selling_Price']]
````

````python
print(X)
````

````python
print(y)
````

````python
# set up our model
linear = LinearRegression()

# train the model
linear.fit(X, y)
````

````python
import matplotlib.pyplot as plt

y_pred = linear.predict(X)
plt.plot(X, y_pred, color='red')

plt.scatter(X, y)
plt.xlabel('Age')
plt.ylabel('Selling_Price (lakhs)')
plt.show()
````

Mean squared error (MSE): This is the most common metric for evaluating the performance of linear regression models. It is calculated by averaging the squared differences between the predicted values and the actual values.

````python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_pred, y)
print("MSE:", mse)
````
